```python
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}
plt.plot('data_x', 'data_y', data=data_dict, label = 'Prive($)')
plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})
plt.legend()
plt.show()

#그래프 영역에 범례를 나타내려면 plot함수에 label 문자열을 저장하고 legend() 함수를 호출한다.
```

```python
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}
plt.plot('data_x', 'data_y', data=data_dict, label = 'Prive($)')
plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})
plt.legend(loc=10)
plt.show()
```

```python
data_1 = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}
data_2 = {'data_x':[1,2,3,4,5], 'data_y':[3,5,9,7,7]}

plt.plot('data_x', 'data_y', data=data_1, label = 'Price($)')
plt.plot('data_x', 'data_y', data=data_2, label = 'Demand(#)')

plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})
plt.legend(loc=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0,2,0.2)

plt.plot(a,a,'r--',a,a**2,'bo',a, a**3,'g-.')

plt.show()
```

```python
a = np.arange(0,2,0.2)

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo', markersize=20)
plt.plot(a, a**3,'g-.', linewidth=10)

plt.show()
```

```python
a = np.arange(0,2,0.2)

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo', markersize=20)
plt.plot(a, a**3,'g-.', linewidth=10)

plt.grid(True)

plt.show()
```

```python
a = np.arange(0,2,0.2)

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo', markersize=20)
plt.plot(a, a**3,'g-.', linewidth=10)

plt.grid(True, axis='y')

plt.show()
```

```python
# 타이틀 제목 설정

a = np.arange(0,2,0.2)

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo', markersize=20)
plt.plot(a, a**3,'g-.', linewidth=10)

plt.grid(True, axis='y',color='blue', alpha=0.1)
plt.title('grap name', loc='right')

plt.show()

```

```python
# 막대 그래프
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.bar(x,values)
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.bar(x,values, width=0.6, align='edge', color='green')
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.barh(x,values, color='green')
plt.yticks(x, years)

plt.show()
```

```python
import numpy as np


np.random.seed(1) #머신러닝, 인공지능에 중요 (동일한 패턴 고정)

n = 50

x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

area = (30*np.random.rand(n)) **2
colors = np.random.rand(n)

fig = plt.figure(figsize=(8,8)) #fig는 액자를 그린다
ax = fig.add_subplot(111, projection='3d') # ax는 사진을 붙힌다
ax.scatter(x,y,z, s=area, c=colors, alpha=0.4, cmap='Set1')


plt.show()
```
